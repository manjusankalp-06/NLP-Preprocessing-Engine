# 2. What happens if stopwords are not removed?
If stopwords are not removed:

There will be many unnecessary words.
It increases processing time.
It may reduce accuracy of the model.

# 3. When removing stopwords can be harmful?
In sentiment analysis
Example: "I am not happy" → removing "not" changes meaning
In questions
Example: "What is your name?" → removing "what" makes no sense

# 4. Stemming vs Lemmatization
Stemming cuts words into root form but may not be correct
Example: "playing" → "play"
Lemmatization gives proper root word using meaning
Example: "better" → "good"

# Preprocessing Function


```python
#task 2
import re

def preprocess_text(text):
    
    # check empty input
    if text == "" or not isinstance(text, str):
        return [], ""
    
    # convert to lowercase
    text = text.lower()
    
    # remove urls and emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)
    
    # remove numbers
    text = re.sub(r"[0-9]", "", text)
    
    # remove repeated characters
    text = re.sub(r'(.)\1+', r'\1', text)
    
    # remove special characters and emojis
    text = re.sub(r'[^a-z\s]', '', text)
    
    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # split into words
    words = text.split()
    
    # remove small words except "no" and "not"
    final_words = []
    for w in words:
        if len(w) > 2 or w == "no" or w == "not":
            final_words.append(w)
    
    clean_sentence = " ".join(final_words)
    
    return final_words, clean_sentence
```

```python
#task 3
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for s in sentences:
    tokens, clean = preprocess_text(s)
    
    print("\nOriginal:", s)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean)
    
    results.append((tokens, clean))
```

```python
#task 4
for i in range(len(results)):
    tokens = results[i][0]
    
    total = len(tokens)
    unique = len(set(tokens))
    
    if total > 0:
        avg_len = sum(len(w) for w in tokens) / total
    else:
        avg_len = 0
    
    print("\nSentence", i+1)
    print("Total tokens:", total)
    print("Unique tokens:", unique)
    print("Average length:", round(avg_len, 2))
```

```python
#task 5
from collections import Counter

all_words = []

for tokens, _ in results:
    all_words.extend(tokens)

count = Counter(all_words)

print("\nTop 10 frequent words:")
print(count.most_common(10))

print("\nLeast 5 frequent words:")
print(count.most_common()[-5:])
```
